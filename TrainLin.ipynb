```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# Generate random data to learn
num_instances = 100
X = 2 * np.random.rand(num_instances,1)
y = 4 + 3 * X + np.random.randn(num_instances,1)
```

```python
plt.plot(X, y, "b.")
plt.xlabel("$x_1$")
plt.ylabel("$y$")
plt.show()
```

```python
# We solve using the normal equations
# First we append a 1 to each instance
X_b = np.c_[np.ones((num_instances,1)), X]

# Then just use the normal equation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
print(theta_best)
```

```python
# Plot model's predictions
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)), X_new]
y_predict = X_new_b.dot(theta_best)
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```
